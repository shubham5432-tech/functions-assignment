1. what is difference between function and method in a python.

```python
def add(x,y):
  return x+y

result=add(5,3) # calling the function
print(result)
```

```python
class dog:
  def __init__(self,name):
    self.name=name

  def bark(self):
    return "woof!"

my_dog=dog("Buddy")
sound=my_dog.bark() # calling the method on the object
print(sound)

```

 6. Explain the concept of generator in python and how they are defined.

  ▶ In python, generators are a type of iterator that efficiently produce a sequence of values on demand,rather than storing them all in memory at once. they are defined using a functionthat uses the yield keyword instead of return. This allows the function to pause and resume execution, yielding a value each time it's called, making them ideal for working with large or potentially infinite sequences

  Key concepts:

1. Iterators:-
Generators are a specific tupe of iterator, which is an object that can return the next time in a sequence one at a time.

2. Yield:-
The yield keyword is used to produce a value from the generator. When a generator function is called, it returns a generator object that can be iterated over using nect() or a loop.

3. Lazy evaluation:-
Generators produce values only when they are requested , making them memory-efficient, especially for large datasets or infinite sequences.

4. Pausing and resuming:-
Generators can be paused and resumed at the yield statement, allowing them to produce values incrementally.





```python
def my_generator(n):
  value=0
  while value < n:
    yield value
    value += 1

#Iterate over the generator
for number in my_generator(5):
  print(number)
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python ?

 ▶ map(), filter() and reduce() are buit in functions in python that operate on iterables(like lists,tuples etc) by applying a function to the items with them.

 map():

 --> Applies a given function to each item in an iterable and returns a new iterator and returns a new iterator with results. It transforms each element individually, maintaining the original structure's length.

 filter():

  --> Creates a new iterator containing only the items from the original iterable for which a given function returns True. It effectively selects elements based on condition

  reduce():

  -->Applies a function cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value. It combines elements sequentially, resulting in a single output. It needs to be imported from functools module.

     

```python
from functools import reduce

numbers = [1,2,3,4,5]

# map:double each other
squared_numbers = list (map(lambda x: x**2, numbers)) #output [1,4,9,16,25]

# filter: get even numbers
even_numbers=list(filter(lambda x: x%2==0,numbers)) #output [2,4]

# reduce : sum all numbers
sum_of_numbers= reduce(lambda x,y: x + y, numbers) #output 15
```

4.Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n ** 0.5) + 1):  # Check up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Example: Check prime numbers from 1 to 200
if __name__ == "__main__":
    print("Prime numbers from 1 to 200:")
    for num in range(1, 201):
        if is_prime(num):
            print(num, end=" ")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

# Example usage:
if __name__ == "__main__":
    n = int(input("Enter number of terms: "))
    fib = FibonacciIterator(n)
    print("Fibonacci sequence:")
    for num in fib:
        print(num, end=" ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

# Example usage
if __name__ == "__main__":
    max_exp = int(input("Enter the maximum exponent: "))
    print(f"Powers of 2 from 2^0 to 2^{max_exp}:")
    for value in powers_of_two(max_exp):
        print(value, end=' ')
```

7. Implement a generator function that reads a file line by line and yields each line as a string

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')

# Example usage
if __name__ == "__main__":
    path = input("Enter file path: ")
    print("File contents:")
    for line in read_file_line_by_line(path):
        print(line)

```
